```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def diode_current(V_in, I_s = 10e-12, V_t=0.0258, R_d=200):
    """Estimate the current accross a diode.
    
    Arguements:
        V_in: The input voltage in volts.
        I_s: Saturation current of the diode at room temperature in Amps.
        V_t: Thermal voltage in Volts
        R_d: Resistance of resistor placed in series.
        
    Returns:
        The estimated current accross the diode.
    """
#     I_s = 10e-12  # Saturation current for silicon diode at room temperature
#     V_t = 0.0258  # Thermal Voltage in Volts
#     R_d = 200 # Ohms
    if V_in <= V_t:
        return 0
    else:
        return V_in / R_d
```

```python
def create_sine_wave(periods=1, amplitude=1, smoothness=100):
    t = np.linspace(0, periods, periods * smoothness + 1)
    sine_wave = amplitude * np.sin(t * 2 * np.pi)
    
    return t, np.array(sine_wave)
```

```python
t, sine_wave = create_sine_wave(periods=2, amplitude=12)
print(t, sine_wave)
```

```python
plt.plot(t, sine_wave)
```
