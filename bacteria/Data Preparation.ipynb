```python
df_32ms = pd.concat([df['wavelength'], df.filter(regex='32ms')], axis=1)
df_32ms.columns = df_32ms.columns.str.replace("_32ms", "")
df_32ms.describe()
```

```python
print(df_32ms.columns.values)
```

```python
growth_stages = ['lag', 'log', 'stat']
```

```python
bacteria_prefixes = np.unique(list(map(lambda col_name: col_name[:2], df_32ms.columns.values[1:])))
bacteria_prefixes
```

```python
df_log = pd.concat([df_32ms['wavelength'], df_32ms.filter(regex='log')], axis=1)
df_log.columns = df_log.columns.str.replace("_log", "")
df_log.describe()
```

```python
species_dfs = []

for species in bacteria_prefixes:
    species_df = df_log.filter(regex=species)
    species_df.columns = species_df.columns.str.replace(species, "")
    species_df.columns = pd.MultiIndex.from_product([[species], species_df.columns])
    
    species_dfs.append(species_df)
    
df_log_multilevel = pd.concat(species_dfs, axis=1)
df_log_multilevel.head()
```

```python
pd.MultiIndex.from_product([['lag']] + df_log_multilevel.columns.levels)
```

```python
growth_stage_dfs = []

for growth_stage in growth_stages:
    growth_stage_df = df_32ms.filter(regex=growth_stage)
    
    species_dfs = []

    for species in bacteria_prefixes:
        species_df = growth_stage_df.filter(regex=species)
        species_df.columns = species_df.columns.str.replace(species, "")
        species_df.columns = species_df.columns.str.replace("_%s" % growth_stage, "")
        species_df.columns = pd.MultiIndex.from_product([[species], species_df.columns])

        species_dfs.append(species_df)

    species_multilevel_df = pd.concat(species_dfs, axis=1)
        
    growth_stage_dfs.append(species_multilevel_df)
    
growth_stages_df = pd.concat(growth_stage_dfs, axis=1, keys=growth_stages)
growth_stages_df.head()
```

```python
growth_stages_df.describe()
```

```python
growth_stages_df.to_csv('data/growth_stages.csv')
df_32ms['wavelength'].to_csv('data/wavelengths.csv')
```
